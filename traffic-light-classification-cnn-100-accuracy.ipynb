```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import models , layers
import tensorflow_hub as hub
import cv2
```

```python
img_width = 220
img_height = 220
batch_size = 120
```

```python
train_gen = ImageDataGenerator(
    rescale = 1./255 , 
    dtype=tf.float32 ,

)
```

```python
train_data = train_gen.flow_from_directory(
     "../input/carla-traffic-lights-images/traffic_light_data/train" ,
    target_size = (img_height , img_width) ,
    color_mode = "rgb" ,
    class_mode='categorical',
    batch_size=batch_size, 
    shuffle=True,
    
)
```

```python
valid_gen = ImageDataGenerator(
    rescale = 1./255 , 
)
```

```python
valid_data = valid_gen.flow_from_directory(
    "../input/carla-traffic-lights-images/traffic_light_data/val" ,
    target_size = (img_height , img_width) ,
    color_mode = "rgb" ,
    class_mode='categorical',
    batch_size=batch_size, 
    shuffle=True,
    
)
```

```python
from tensorflow.keras.applications import VGG16
```

```python
VGG16=VGG16(include_top=False,weights='imagenet',input_shape=(img_height,img_width,3))
```

```python
VGG16.trainable=False
```

```python
from keras.layers import Dense ,Flatten ,Conv2D ,MaxPooling2D ,Dropout ,BatchNormalization
```

```python
model = keras.Sequential(
[
    VGG16,
    MaxPooling2D(2,2) ,
    Flatten() , 
    Dense(280 , activation ="relu") ,
    Dropout(0.2) , 
    Dense(4 , activation ="softmax")
    
]
)
```

```python
model.summary()
```

```python
model.compile(optimizer = "adam" , loss ="categorical_crossentropy" , metrics = ["accuracy"])
```

```python
from tensorflow.keras.callbacks import Callback, EarlyStopping
```

```python
my_callbacks  = [EarlyStopping(monitor='val_loss',
                              min_delta=0,
                              patience=2,
                              mode='auto')]
```

```python
history = model.fit(train_data , validation_data = valid_data , epochs = 20 , callbacks = my_callbacks )
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
epochs = range(len(acc))

plt.plot(epochs, acc, 'b', label='Training accuracy')
plt.plot(epochs, val_acc, 'r', label='Validation accuracy')
plt.title('Training accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'b', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Training loss')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from tensorflow.keras.preprocessing import image as image_utils
from tensorflow.keras.applications.imagenet_utils import preprocess_input

def show_image(image_path):
    image = mpimg.imread(image_path)
    plt.imshow(image)

def make_predictions(image_path):
    show_image(image_path)
    image = image_utils.load_img(image_path, target_size=(img_height, img_width))
    image = image_utils.img_to_array(image)
    image = image.reshape(1,img_height,img_width,3)
    image = preprocess_input(image)
    preds = model.predict(image)
    return preds
```
